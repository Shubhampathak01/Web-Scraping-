```python
import selenium 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver= webdriver.Chrome()
driver.get('https://www.google.com/')
```

```python
url= 'https://www.naukri.com/'
driver.get(url)
```

```python
designation= driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Scientist')
```

```python
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Noida')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
Designation = []
Company=[]
Skill=[]
Address=[]
Exeprience=[]
Qualification=[]
```

```python
try:
    designtion=driver.find_elements(By.XPATH,'//a[@class="title "]')
    for i in designtion:
        Designation.append(i.text)
except NoSuchElementException:
    Designation.append('_')
except StaleElementReferenceException:
    Designation.append('_')
    
try:
    company=driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')
    for i in company:
        Company.append(i.text)
except NoSuchElementException:
    Company.append('_')
except StaleElementReferenceException:
    Company.append('_')
try:
    skill=driver.find_elements(By.XPATH,'//div[@class=" row5"]')
    for i in skill:
        Skill.append(i.text)
except NoSuchElementException:
    Skill.append('_')
except StaleElementReferenceException:
    Skill.append('_')

try:
    address= driver.find_elements(By.XPATH,'//span[@class="loc-wrap ver-line"]')
    for i in address:
        Address.append(i.text)
except NoSuchElementException:
    Address.append('_')
except StaleElementReferenceException:
    Address.append('_')
try:
    exeprience=driver.find_elements(By.XPATH,'//span[@class="exp-wrap"]')
    for i in exeprience:
        Exeprience.append(i.text)
except NoSuchElementException:
    Exeprience.append('_')
except StaleElementReferenceException:
    Exeprience.append('_')
try:
    qulification=driver.find_elements(By.XPATH,'//div[@class=" row4"]')
    for i in qulification:
        Qualification.append(i.text)
except NoSuchElementException:
    Qualification.append('_')
except StaleElementReferenceException:
    Qualification.append('_')
    
```

```python
a = {'Name_Des.': Designation,'COMPANY': Company,'SKILL_SET':Skill,'OFFICE_AD.': Address, 'EXEPRIENCE':Exeprience, 'DESCRIPTION':Qualification}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
print("JOB IN NOIDA SECTOR FOR DATA SCIENCE")
df
```
